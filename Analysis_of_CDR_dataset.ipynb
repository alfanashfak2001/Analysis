```python
#import needed packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import csv
```

```python
#save the original dataset as backup
Original_data = pd.read_csv('CDR-Call-Details.csv')
```

```python
#duplicate original data as data
data=Original_data
```

```python
# Find duplicates
duplicates = data[data.duplicated('Phone Number')]
duplicates.sum()
```

```python
# Remove duplicates
data.drop_duplicates(subset='Phone Number', inplace=True)
```

```python
data.columns
```

```python
#create a new column by adding three column values for total minutes
data['Local call minutes'] = data['Eve Mins'] + data['Night Mins'] + data['Day Mins']
data['Local call charge'] = data['Eve Charge'] + data['Night Charge'] + data['Day Charge']
data['Local call count'] = data['Eve Calls'] + data['Night Calls'] + data['Day Calls']
```

```python
# Load your dataset into a pandas DataFrame
df = pd.read_csv('Cleaned_CDR_data.csv')

# create a new column Intlpackageuser by the value greater than 0 of the international minutes
df['IntlPackageuser'] = df['Intl Mins'].apply(lambda x: 1 if x>0 else 0)

# create a new column Localpackageuser by the value greater than 0 of the local minutes minutes
df['LocalPackageuser'] = df['Local call minutes'].apply(lambda x: 1 if x>0 else 0)

#uniqe value details
df['IntlPackageuser'].value_counts()
df['LocalPackageuser'].value_counts()
```

```python
df.head(5)
```

```python
# Count the number of users for each combination
users_with_both_packages = ((df['IntlPackageuser'] == 1) & (df['LocalPackageuser'] == 1)).sum()
users_with_Localpackage_only = ((df['LocalPackageuser'] == 1) & (df['IntlPackageuser'] == 0)).sum()

# Create the grouped bar plot
options = ['users_with_both_packages', 'users_with_Localpackage_only']
user_counts = [users_with_both_packages,users_with_Localpackage_only]
plt.figure(figsize=(8,20))
plt.bar(options, user_counts)
plt.xlabel('Options')
plt.ylabel('Number of Users')
plt.title('Usage of International package and Local package by Users')
plt.show()
print("users_with_both_packages ",users_with_both_packages.sum())
print("users_with_Localpackage_only ",users_with_Localpackage_only.sum())
```
